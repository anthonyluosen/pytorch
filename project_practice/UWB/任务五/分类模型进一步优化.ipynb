```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import tensorflow as tf 
from tensorflow.keras import backend as b
from tensorflow.keras.models import Sequential 
from tensorflow.keras.layers import Activation
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.metrics import categorical_crossentropy
from sklearn.model_selection import train_test_split
```

```python
import os
os.listdir()
# data = pd.read_csv()
```

```python
data = pd.read_csv('搞过的数据.csv')
data.head()
```

```python
data.drop('Unnamed: 0',axis=1,inplace = True)
data.head()
```

```python
X = data.drop(['X','Y','Z'],axis=1)
y = data[['X','Y','Z']]

x_train , x_test , y_train , y_test = train_test_split(X , y , train_size=0.8 )

n_inputs = x_train.shape[1]
model = Sequential([
    Dense(n_inputs , input_shape=(n_inputs, ) , activation = 'relu'),
    Dense(128 , activation = 'relu'),
    Dense(256 , activation = 'relu'),
    Dense(256 , activation = 'relu'),
    Dense(512 , activation = 'relu'),
    Dense(3)
])
```

```python
print(X.shape)
print(y.shape)
model.summary()
```

```python
model.compile(Adam(lr=0.0001) , loss = 'mean_squared_error')
model.fit(x_train , y_train , validation_split=0.2 ,  epochs=1000 ,
          batch_size=64 ,shuffle=True , verbose=2
         )
```

```python
model.save('第三次拟合算法_求位置.h5')
```

```python
X = data.drop(['X','Y','Z','los','nlos'],axis=1)
y = data['nlos']

x_train , x_test , y_train , y_test = train_test_split(X , y , train_size=0.8 )

n_inputs = x_train.shape[1]
model = Sequential([
    Dense(n_inputs , input_shape=(n_inputs, ) , activation = 'relu'),
    Dense(128 , activation = 'relu'),
    Dense(256 , activation = 'relu'),
    Dense(256 , activation = 'relu'),
    Dense(512 , activation = 'relu'),
    Dense(2 , activation = 'softmax')
])
```

```python
print(X.shape)
print(y.shape)
model.summary()
```

```python
model.compile(Adam(lr=0.0001) , loss = 'sparse_categorical_crossentropy',
             metrics = ['accuracy'])
model.fit(x_train , y_train , validation_split=0.2 ,  epochs=1000 ,
          batch_size=64 ,shuffle=True , verbose=2
         )
```
